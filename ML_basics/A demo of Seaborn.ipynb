## A Data Visualisation Experiment with Seaborn

```python
import seaborn as sns
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

We demonstrate the styles of plot using the dataset from tips. sns.set() can set 5 styles of plotting: darkgrid, whitegrid, dark, white and ticks. The default is darkgrid.

```python
sns.set()
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size', data=tips)
```

```python
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size', data=tips)
```

We can change the colour of the points using palette.

```python
sns.set(palette='RdBu')
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size', data=tips)
```

### A mixture of points and lines using relplot():

```python
sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill',y='tip',hue='day', data=tips)
plt.show()
```

```python
# Add column to the arguments
sns.relplot(x='total_bill',y='tip',hue='day', data=tips, col='sex')
plt.show()
```

```python
# Add row to arguments
sns.relplot(x='total_bill',y='tip',hue='day', data=tips, col='sex',row='time')
plt.show()
```

### Line plot with kind='line'

```python
fmri = sns.load_dataset('fmri')
sns.relplot(x='timepoint', y='signal', hue='event', style='event',col='region', kind='line', data=fmri)
plt.show()
```

### Linear Regression with lmplot()

```python
sns.lmplot(x='total_bill', y ='tip', data=tips)
plt.show()
```

```python
# Setting confidence interval of the linear regression line, default is 95
sns.lmplot(x='total_bill', y='tip', data=tips, ci=60)
plt.show()
```

```python
# Separate into two dataset by hue
sns.lmplot(x='total_bill', y='tip', data=tips, ci=60, hue='smoker')
plt.show()
```

```python
# Nonlinear regression with lowess
sns.lmplot(x='total_bill', y='tip', data=tips, lowess=True)
plt.show()
```

### Classified Scatter Plot with stripplot()

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.show()
```

### Violinplot() to demonstrate unimodal or multimodal

```python
sns.violinplot(x='total_bill', y='day', hue='time', data=tips)
plt.show()
```
